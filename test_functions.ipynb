```python
def Qout_function(w1, w2, w3):
    if w1<= 600:
        q_1 = 3.2216 + 0.08378681 * w1
    else:
        q_1=  3.22 + 0.083 * 600 + 0.8371 * (w1 - 600)

    if w2<= 600:
        q_2 = 3.2216 + 0.08378681 * w2
    else:  
        q_2 = 3.22 + 0.083 * 600 + 0.8371 * (w2 - 600)

    if w3<= 600:
        q_3 = 3.2216 + 0.08378681 * w3
    else:  
        q_3 = 3.22 + 0.083 * 600 + 0.8371 * (w3 - 600)
    
    q_tot = q_1+q_2+q_3
    return q_tot
```
